```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns

%pylab inline
```

```python
df = pd.read_csv("seattle-weather.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
# Все данные корректно заполнены
df.isnull().sum()
```

```python
# Сначала разделим на признаки (X), и ответы (y)
# предсказывать будем Summary, сам прогноз погоды

y = df['weather']
X = df.drop(['date', 'weather'], axis=1)
```

```python
X.head()
```

```python
y.head()
```

```python
# Теперь можно поделить на тестовую и валидационную выборку в соотношении 0.3
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
print("data:   ", df.shape[0])
print("X_train:", X_train.shape[0])
print("X_test: ", X_test.shape[0])
```

```python
from sklearn.metrics import accuracy_score
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier, RandomForestClassifier
```

```python
rf_clf = RandomForestClassifier(
    criterion="entropy", min_samples_leaf=8, n_jobs=-1
  )
rf_clf.fit(X_train, y_train)
y_pred = rf_clf.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from sklearn.ensemble import GradientBoostingClassifier

gb_clf = GradientBoostingClassifier(
    max_depth=3, learning_rate=0.1, n_estimators=22
  )
gb_clf.fit(X_train, y_train)
y_pred = gb_clf.predict(X_test)

accuracy_score(y_test, y_pred)
```
